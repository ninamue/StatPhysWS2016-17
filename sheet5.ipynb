# Ex. 1 Markov Inequality

```julia
histogram(rand(Normal(0,1),10^5),title="Gaussian distributed random variable"*"\n"* 
"10^5 trials") 
```

```julia
# calculate 
# draw N i.i.d. randdom variables, sum up and divide by N. Repeat M times and store results in vector. 

function calc(N,M)
    tmp=[]
    for i in 1:M
        push!(tmp,sum(rand(Normal(0,1),N))/N) 
    end
    tmp
end
```

```julia
histogram(calc(10,10^5), lab="N=10",title="Sum of N Gaussian random variables, divided by N"*"\n"* 
"10^5 trials")
```

```julia
histogram!(calc(50,10^5), lab="N=50")
```

```julia
histogram!(calc(100,10^5), lab="N=100")
```

```julia
histogram!(calc(1000,10^5), lab="N=1000")
```

```julia
histogram!(calc(10^5,10^5), lab="N=10 000")
```

# Ex.2: Ising Paramagnet

```julia
M(h,T)=tanh(h/T)
```

```julia
plot(linspace(0.1,100,500),T->M(1,T),lab="h=1",xlab="T",ylab="<M>",title="Magnetization as function of temperature")
plot!(linspace(0.1,100,500),T->M(3,T),lab="h=3")
plot!(linspace(0.1,100,500),T->M(5,T),lab="h=5")
plot!(linspace(0.1,100,500),T->M(15,T),lab="h=15")
```

```julia
plot(linspace(0.1,50,500),h->M(h,5),lab="T=5",xlab="h",ylab="<M>",title="Magnetization as function of external field")
plot!(linspace(0.1,50,500),h->M(h,10),lab="T=10")
plot!(linspace(0.1,50,500),h->M(h,15),lab="T=15")
plot!(linspace(0.1,50,500),h->M(h,30),lab="T=30")
```

# What you should have learned: 

strong field/low temperature = ordered configuration. 
Reason: (M=N or M=-N) energetically prefered state, deviations have relatively high energy costs. 

weak field/high temperature = disordered phase. 
Reason: (M=0) entropically favoured, far more microstates with M=0 than M=+N/-N. Deviations have relatively low energy cost.
