# Van der Waals Gas
Setze kB=a=b=1

```julia
using Plots
plotlyjs()
vc=3
Tc=8/27
pc=3/8*Tc/vc
```

```julia
A(p,T)=3*p+1/3-3/2*T
B(T)=9/8*T-1/3
C(T)=(9/8)^2*T-1/3
g(v,T,p)=A(p,T)*v+B(T)*(v^2-v^3)+C(T)*v^4
```

```julia
x=-2:0.01:2

plot(x,x->g(x,Tc,2*pc),lab="5*pc,Tc")
plot!(x,x->g(x,Tc,pc),lab="pc,Tc")
plot!(x,x->g(x,Tc,pc/5),lab="pc/5,Tc")
```

```julia
plotlyjs()
x=-3:0.01:2
plot(x,x->g(x,1.05*Tc,pc),lab="pc,1.05*Tc",ylim=(-0.2,0.2))
plot!(x,x->g(x,Tc,pc),lab="pc,Tc")
plot!(x,x->g(x,0.95*Tc,pc),lab="pc,0.95*Tc")
plot!(x,x->g(x,0.93*Tc,pc),lab="pc,0.90*Tc")
```
